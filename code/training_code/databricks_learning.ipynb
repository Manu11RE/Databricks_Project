## This is the import data from GitHub space

```python
import pandas as pd

# 1. Workspace route
input_path = "/Workspace/Repos/raw_data/Databricks_Project/data/raw/orderscsv.csv"

# 2. Read the data with pandas
pdf = pd.read_csv(input_path)

# 3. Convert the df to spark
df = spark.createDataFrame(pdf)

# 4. Storge the data in a table
target_table = "workspace.github_data.orders_table"
df.write.mode("overwrite").saveAsTable(target_table)

print("Finally!")
```
